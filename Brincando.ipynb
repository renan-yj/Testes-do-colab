# Main

```python
import gymnasium as gym
from IPython import display
from matplotlib import animation
import matplotlib.pyplot as plt

# 1. Cria o ambiente
env = gym.make("CartPole-v1", render_mode="rgb_array")
observation, info = env.reset()
frames = []

# 2. Roda a simulação por mais tempo (ex: 200 passos)
for _ in range(500):
    frames.append(env.render()) # Salva o frame na lista
    action = env.action_space.sample()
    observation, reward, terminated, truncated, info = env.step(action)
    if terminated or truncated:
        observation, info = env.reset()
env.close()

# 3. Cria a animação suave
plt.figure(figsize=(frames[0].shape[1] / 72.0, frames[0].shape[0] / 72.0), dpi=72)
patch = plt.imshow(frames[0])
plt.axis('off')

def animate(i):
    patch.set_data(frames[i])

# Aqui você controla a velocidade (interval = milissegundos entre frames)
anim = animation.FuncAnimation(plt.gcf(), animate, frames=len(frames), interval=30)
display.display(display.HTML(anim.to_jshtml()))
plt.close()
```
